```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
```

## Oppgave 7 a)


```python
# Siden farten er konstant i denne oppgaven vil W = F*d

C=0.24
rho=1.23 #kg/m^3
A=2.33 # m^2
# Største model S masse
M=2250 # kg
d=10000
v_1 = 22.2 # 80 km/h
v_2 = 27.8 # 100 km/h
w_1 = 0.5*rho*A*0.21*v_1*d
w_2 = 0.5*rho*A*0.21*v_2*d
print(f"Dersom bilen kjører 80 km/h vil luftmotstanden gjøre et arbeid på {w_1*2.77818**-7:.3} kWh")
print(f"Dersom bilen kjører 100 km/h vil luftmotstanden gjøre et arbeid på {w_2*2.77818**-7:.3} kWh")
```

## Oppgave 7 b)

```python
def v(t,vmax):
    v=0.5*vmax*1000/3600 *(np.sin((np.pi/10)*t-np.pi/2)+1)
    return v

# Lager et intervall fra 0 til 10 som kan brukes som t i funksjonen v.
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y >= 0) & (y <= 10)]
y.sort()
x = np.arange(len(y))

func_1 = v(x, 80)
func_2 = v(x, 100)

plt.xlim(0, 10)
plt.plot(func_1)
plt.plot(func_2)
plt.xlabel("Tid (t)")
plt.ylabel("Fart (m/s)")
plt.legend(["Ved maksfarten 80 km/h", "Ved maksfarten 100 km/h"])
plt.show()
```

## Oppgave 7 c)

```python
result_80,err = integrate.quad(v, 0, 10,args=80)
result_100,err = integrate.quad(v, 0, 10,args=100)

print(f"Bilen som har maksfart 80 km/h kjører {result_80:.4} m på de første 10 sekundene")
print(f"Bilen som har maksfart 100 km/h kjører {result_100:.4} m på de første 10 sekundene")
```

## Oppgave 7 e)

```python
def drag_effekt(t,vmax):
    return 0.5*A*rho*C*v(t,vmax)**3

friksjon_80,err = integrate.quad(drag_effekt, 0, 10, args=80)
friksjon_100,err = integrate.quad(drag_effekt, 0, 10, args=100)

print(f"Etter 10 sekunder har friksjonen gjort {friksjon_80:.6} J arbeid i akselerasjonsfasen for bilen med 80 km/h som maksfart")
print(f"Etter 10 sekunder har friksjonen gjort {friksjon_100:.6} J arbeid i akselerasjonsfasen for bilen med 100 km/h som maksfart")
```

## Oppgave 7 g)

```python
#P=Fv , F = drag-effekt
C_1 = 0.21

def drag_effekt_ny_C(t,vmax):
    return 0.5*A*rho*C_1*v(t,vmax)**3

effekt_tap_1 = drag_effekt(10, 80) - drag_effekt_ny_C(10, 80)
effekt_tap_2 = drag_effekt(10, 100) - drag_effekt_ny_C(10, 100)

print(f"Effekttapet til bilen med maksfarten 80 km/h blir {effekt_tap_1:.4} W")
print(f"Effekttapet til bilen med maksfarten 100 km/h blir {effekt_tap_2:.4} W")
```
